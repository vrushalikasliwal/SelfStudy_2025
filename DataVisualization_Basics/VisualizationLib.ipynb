-#Histogram

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('Automobile_data.csv')
df.head()
```

```python
df.shape
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
plt.hist(df['width'])
plt.title('Width Freq Table')
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.show()
```

```python
width_mean=df['width'].mean()
width_median=df['width'].median()
df['width'].hist(bins=10)
print(width_mean)
print(width_median)
plt.axvline(width_mean,color='y')
plt.axvline(width_median,color='r')
plt.xlabel('Width')
plt.ylabel('Freq')
plt.show()
```

```python
df['body-style'].hist(bins=5,grid=False,alpha=0.5,histtype='bar',ec='green')
plt.show()
```
