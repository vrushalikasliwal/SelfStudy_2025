```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#method of genrate random data in array: random, randint, arrange, linspace, once, zeros
```

```python
a=np.linspace(10,200,500)
b=np.random.random_sample(500)
print(a)
print(b)
```

```python
plt.scatter(a,b)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(5,5)
plt.rcParams['figure.dpi']=100
```

```python
plt.scatter(a,b)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('A Vs B')
plt.show()
```

```python
plt.scatter(a,b,color='r',edgecolors='b')
plt.show()
```
