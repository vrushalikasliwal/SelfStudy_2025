```python
#Annova Test Fucntion
f,p=sc.f_oneway(df['SepalLengthCm'],df['SepalWidthCm'],df['PetalLengthCm'],df['PetalWidthCm'])
f,p
```

```python
if p<0.05:
    print('Reject Null Hypo! Atleast one sample is diffrent')
else:
    print('Fail to reject null hypo!')
```
