```python
import numpy as np
```

```python
n1=247 # data points for sample1
p1=0.37 #Proportion 1 from sample 1

n2=308 # data points for sample2
p2=.39 #Proportion 2 from sample 2
```

```python
sample1= np.random.binomial(1,p1,n1)
sample1
```

```python
sample2= np.random.binomial(1,p2,n2)
sample2
```

```python
len(sample1), len(sample2)
```

```python
sample1.mean(), sample2.mean()
```

```python
import statsmodels.api as sm
```

```python
t,p, df= sm.stats.ttest_ind(sample1,sample2)
t,p,df
```

```python
p
```

```python
if p < 0.05:
 print('Reject null hypo..')
else:
 print ('Fail to reject null hypo, accept null hypo..')
```
