```python
obs=pd.crosstab(index=df['Athlete'],columns=df['Smoker'])
obs
```

```python
chi2, p,df,exp=sc.chi2_contingency(obs)
chi2,p,df,exp
```

```python
if p<0.05:
    print('Ho: Reject Null Hypo! Cols are independent')
else:
    print('Ha: Fail to reject null hypo! Cols are dependent on each other')
```
