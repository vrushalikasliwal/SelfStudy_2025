```python
### One sample One Tail test

#Example:- 
#An outbreak of salmunella related illness was attributed to IceCream, produced at a certain factory, 
#Scientific measured the level of salmonella in randomly sampled bathces of icecream, the levels (in MPN/gm) were 
#[0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418]
# Is there evidence that the mean level of salmonella in the ice cream is > 0.3MPN/gm

# here sample sice (n)= 9 
# so dgress of freddon (df)=n-1=8

#1---------------------------------------------------
#   H(o): mue (u)<0.3 (No action)
#   H(a): mue (u)>0.3 (Take some action)
#2---------------------------------------------------
# Population SD (sigma=6) not given then use t-Test
# t-Test=(X-mean / mew(U)) / (S/square root of n))
#3---------------------------------------------------
# one Tail test becuase (< and >) in define part
# scipy.stats.ttest_1Samp(sample_data, muew)
# sample_data=[0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418]
# muew= 0.3 (Population mean)
#4---------------------------------------------------
# alpha: CL=95% so alpha=0.05
#5---------------------------------------------------
#  p_Value: ?
#6---------------------------------------------------
#  if p_Vlaue < alpha : Reject null hypothesis
#  if p_Vlaue > alpha : Accept Null Hypothesis
```

```python
from scipy import stats
import scipy
```

```python
s_Data= [0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418]
s_Data
```

```python
muew= 0.3 # population mean
```

```python
scipy.stats.ttest_1samp(s_Data,muew)
```

```python
s,p=scipy.stats.ttest_1samp(s_Data,muew) 
s
```
