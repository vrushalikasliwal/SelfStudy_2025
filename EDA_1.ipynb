```python
def outlires(data,col):
    q1=data[col].quantile(0.25)
    q3=data[col].quantile(0.75)
    iqr=q3-q1
    lwr=q1-(1.5*iqr)
    upr=q3-(1.5*iqr)
    return lwr,upr
```

```python
outlires(df,'Ozone')
```

```python
outlires(df,'Temperature')
```

```python
df.columns
```

```python
outlires(df,'Wind')
```

```python
#Replace outliers with UE OR LE
def outliers_detect(data, allCol):
    for col in allCol:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        iqr = q3 - q1
        lwr = q1 - (1.5 * iqr)
        upr = q3 + (1.5 * iqr)   # corrected upper bound

        data[col] = np.where(
            data[col] > upr, upr,
            np.where(data[col] < lwr, lwr, data[col])
        )
    
```

```python
outliers_detect(df,df.drop(columns=['Weather']).columns)
```
